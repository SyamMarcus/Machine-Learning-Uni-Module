```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
col = ["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","target"
]
df = pd.read_csv("/home/shaun/Desktop/Work/Sem1/6006CEM_ML/CW/Datasets/heart.csv",names=col)
df
```

```python
x = df.iloc[:,:-1].values  # set the x values to the first 13 collumns
y = df.iloc[:,-1:].values  # set the y values to the last collumn only
#print(x)
#print(y)
#print(x.shape)
#print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)  # split the x and y data in to training and test sets
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)  # set the training model with increased max_iter
model.fit(x_train,y_train)  # build the model using the training data
```

```python
y_train_pred = model.predict(x_train)  # compare the training data with the test data
y_test_pred = model.predict(x_test)
accuracy = model.score(x_test, y_test)  
print(accuracy)
```
